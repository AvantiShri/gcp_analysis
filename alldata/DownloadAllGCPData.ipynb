Iterate through all the days and download the data.

Note: at the time of writing, the gcp data download page (https://web.archive.org/web/20240213150836/https://global-mind.org/wget.html) suggests that the data for a year can be downloaded with commands of the form `wget -N -r -nH --cut-dirs=1 --limit-rate=125k https://global-mind.org/data/eggsummary/wget/wget2010.html`. However, in doing this, I discovered that the `wget[year].html` page for some years was incomplete, for example https://web.archive.org/web/20240615065650/https://global-mind.org/data/eggsummary/wget/wget2018.html only lists days up till `2018-12-02` even though more days are available. Thus, instead I just iterate through all the days in order to make sure I did not miss any days.

```python
import datetime

%cd /content/drive/MyDrive/GCP_data/all_data

for year in range(1998,2024):
  current_day = datetime.datetime(year=year, month=1, day=1)
  while current_day.year==year:
    string_date = current_day.strftime("%Y-%m-%d")
    str_year = str(year)
    print(string_date)
    !wget -N -r -nH --cut-dirs=1 --limit-rate=125k https://global-mind.org/data/eggsummary/{str_year}/basketdata-{string_date}.csv.gz
    current_day = current_day + datetime.timedelta(days=1)
```
